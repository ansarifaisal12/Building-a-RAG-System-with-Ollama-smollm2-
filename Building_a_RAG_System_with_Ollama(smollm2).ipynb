```python
import os
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.llms import Ollama
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from typing import Dict, List
import warnings
warnings.filterwarnings('ignore')
```

# Financial prompt templates

```python
# Financial prompt templates
FINANCIAL_ANALYSIS_TEMPLATE = """
You are a financial expert analyzing documents. Use the following context to answer the question.
Follow these guidelines:
1. Focus on quantitative data and financial metrics
2. Highlight key financial indicators and trends
3. Consider market conditions and economic factors
4. Provide balanced risk assessment when relevant
5. Use financial terminology appropriately
6. Cite specific numbers and data points from the context

Context: {context}
Question: {question}

Provide a structured analysis following these steps:
1. Key Financial Metrics
2. Trend Analysis
3. Risk Assessment
4. Recommendations (if applicable)

Analysis:
"""

RISK_ASSESSMENT_TEMPLATE = """
As a risk analyst, evaluate the following information and provide a risk assessment.
Consider these risk categories:
- Market Risk
- Credit Risk
- Operational Risk
- Liquidity Risk
- Regulatory Risk

Context: {context}
Question: {question}

Provide your analysis in this format:
1. Risk Identification
2. Risk Quantification
3. Mitigation Strategies
4. Overall Risk Rating

Analysis:
"""

MARKET_ANALYSIS_TEMPLATE = """
As a market analyst, analyze the following information and provide market insights.
Focus on:
- Market trends
- Competitive analysis
- Growth opportunities
- Market challenges
- Industry benchmarks

Context: {context}
Question: {question}

Structure your response as:
1. Market Overview
2. Competitive Position
3. Growth Analysis
4. Strategic Recommendations

Analysis:
"""
```

#RAG Pipeline


```python
class FinanceRAG:
    def __init__(self):
        self.llm = Ollama(model="smollm2")
        self.embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

    def load_documents(self, pdf_path: str) -> List:
        loader = PyPDFLoader(pdf_path)
        return loader.load()

    def split_text(self, documents: List) -> List:
        splitter = RecursiveCharacterTextSplitter(
            chunk_size=1000,
            chunk_overlap=200,
            length_function=len,
        )
        return splitter.split_documents(documents)

    def setup_vector_store(self, chunks: List) -> Chroma:
        return Chroma.from_documents(
            documents=chunks,
            embedding=self.embeddings,
            persist_directory="finance_db"
        )

    def create_specialized_chain(self, vectorstore: Chroma, analysis_type: str) -> RetrievalQA:
        # Select appropriate template based on analysis type
        template_map = {
            'financial': FINANCIAL_ANALYSIS_TEMPLATE,
            'risk': RISK_ASSESSMENT_TEMPLATE,
            'market': MARKET_ANALYSIS_TEMPLATE
        }

        prompt_template = PromptTemplate(
            template=template_map.get(analysis_type, FINANCIAL_ANALYSIS_TEMPLATE),
            input_variables=["context", "question"]
        )

        chain = RetrievalQA.from_chain_type(
            llm=self.llm,
            chain_type="stuff",
            retriever=vectorstore.as_retriever(),
            chain_type_kwargs={"prompt": prompt_template},
            return_source_documents=True
        )

        return chain

    def query_document(self, chain: RetrievalQA, query: str, analysis_type: str) -> Dict:
        # Add specific prefixes based on analysis type
        prefix_map = {
            'financial': "Provide a financial analysis of: ",
            'risk': "Assess the risks related to: ",
            'market': "Analyze the market aspects of: "
        }

        prefixed_query = f"{prefix_map.get(analysis_type, '')}{query}"
        return chain({"query": prefixed_query})
```

```python
def process_financial_document(pdf_path: str) -> Dict:
    rag = FinanceRAG()

    print("Loading and processing document...")
    documents = rag.load_documents(pdf_path)
    chunks = rag.split_text(documents)
    vectorstore = rag.setup_vector_store(chunks)

    # Create different analysis chains
    financial_chain = rag.create_specialized_chain(vectorstore, 'financial')
    risk_chain = rag.create_specialized_chain(vectorstore, 'risk')
    market_chain = rag.create_specialized_chain(vectorstore, 'market')

    return {
        'rag': rag,
        'financial_chain': financial_chain,
        'risk_chain': risk_chain,
        'market_chain': market_chain
    }

def analyze_document(components: Dict, query: str, analysis_type: str = 'financial') -> str:
    chain_map = {
        'financial': components['financial_chain'],
        'risk': components['risk_chain'],
        'market': components['market_chain']
    }

    chain = chain_map.get(analysis_type)
    response = components['rag'].query_document(chain, query, analysis_type)
    return response['result']
```

```python
from google.colab import files
uploaded = files.upload()
pdf_filename = list(uploaded.keys())[0]

# Process the document
components = process_financial_document(pdf_filename)
```

```python
financial_query = "What are the key financial metrics and their trends?"
risk_query = "What are the main risk factors identified in the document?"
market_query = "What are the market opportunities and competitive advantages?"

# Get different types of analysis
financial_analysis = analyze_document(components, financial_query, 'financial')
risk_analysis = analyze_document(components, risk_query, 'risk')
market_analysis = analyze_document(components, market_query, 'market')

print("Financial Analysis:", financial_analysis)
print("\nRisk Analysis:", risk_analysis)
print("\nMarket Analysis:", market_analysis)
```
